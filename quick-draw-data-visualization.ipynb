```python
import os
import re
import json
import random
import urllib.request
import numpy as np
import matplotlib.pyplot as plt

from tqdm import tqdm
from pathlib import Path
from matplotlib.gridspec import GridSpec
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
random.seed(42)
np.random.seed(42)
tf.random.set_seed(42)
```

```python
DATA_DIR = Path("data")
RAW_DIR = DATA_DIR / "raw"
MODEL_DIR = Path("models")
```

```python
NUM_CLASSES = 50
MAX_SAMPLES_PER_CLASS = 40000
BATCH_SIZE = 512
EPOCHS = 20
```

```python
def load_class_data(class_names, max_samples_per_class=10000):
    class_data = {}
    
    for class_name in tqdm(class_names, desc="Loading data"):
        try:
            file_path = RAW_DIR / f"{class_name}.npy"
            data = np.load(file_path)
        
            if len(data) > max_samples_per_class:
                indices = np.random.choice(len(data), max_samples_per_class, replace=False)
                data = data[indices]
            
            class_data[class_name] = data
            
        except Exception as e:
            print(f"Error loading {class_name}: {e}")
    
    return class_data
```

```python
def select_classes():
    path = DATA_DIR / 'selected_classes.json'
    
    with path.open('r', encoding='utf-8') as f:
        selected_classes = json.load(f)

    return(selected_classes)
```

```python
class_data = load_class_data(select_classes())
class_data
```

```python
def visualize_class_examples(class_data, samples_per_class=5):
    num_classes = len(class_data)
    
    # Create a figure with a grid of subplots
    fig = plt.figure(figsize=(15, num_classes * 1.5))
    gs = GridSpec(num_classes, samples_per_class + 1, figure=fig)
    
    # For each class, show the class name and some examples
    for i, (class_name, data) in enumerate(class_data.items()):
        # Class name in the first column
        ax = fig.add_subplot(gs[i, 0])
        ax.text(0.5, 0.5, class_name, ha='center', va='center', fontsize=10)
        ax.axis('off')
        
        # Show random examples
        indices = np.random.choice(len(data), samples_per_class, replace=False)
        for j, idx in enumerate(indices):
            ax = fig.add_subplot(gs[i, j + 1])
            ax.imshow(data[idx].reshape(28, 28), cmap='gray')
            ax.axis('off')
    
    plt.tight_layout()
    plt.show()
```

```python
print("\nVisualizing examples from each class...")
visualize_class_examples(class_data)
```

```python
def visualize_normalized_data(class_data):
    # Select a few classes and samples
    selected_classes = list(class_data.keys())[:3]
    fig, axes = plt.subplots(3, 3, figsize=(10, 10))
    
    for i, class_name in enumerate(selected_classes):
        data = class_data[class_name]
        idx = np.random.choice(len(data))
        sample = data[idx]
        
        # Original
        axes[i, 0].imshow(sample.reshape(28, 28), cmap='gray')
        axes[i, 0].set_title(f"{class_name} - Original")
        axes[i, 0].axis('off')
        
        # Normalized (0-1)
        normalized = sample.astype('float32') / 255.0
        axes[i, 1].imshow(normalized.reshape(28, 28), cmap='gray')
        axes[i, 1].set_title(f"{class_name} - Normalized")
        axes[i, 1].axis('off')
        
        # Augmented (flipped)
        flipped = np.flip(sample.reshape(28, 28), axis=1)
        axes[i, 2].imshow(flipped, cmap='gray')
        axes[i, 2].set_title(f"{class_name} - Flipped")
        axes[i, 2].axis('off')
    
    plt.tight_layout()
    plt.show()
```

```python
print("\nVisualizing normalized data...")
visualize_normalized_data(class_data)
```
